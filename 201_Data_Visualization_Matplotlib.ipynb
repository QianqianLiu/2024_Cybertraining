**2. Barplot**

```python
zones = ['Pacific', 'Atlantic', 'Indian', 'Arctic', 'Southern']
average_depths = np.array([4282, 3926, 3963, 1205, 4000])
plt.figure(figsize=(10, 6))
bars = plt.bar(zones, average_depths, color=['blue', 'orange', 'green', 'red', 'purple'])
#bars = plt.bar(zones, average_depths, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']) # hex code of RGB colors
plt.title("Average Depth of World's Oceans")
plt.xlabel('Ocean Name')
plt.ylabel('Average Depth (meters)')

for bar in bars:
  yval=bar.get_height()
  plt.text(bar.get_x()+bar.get_width()/2,yval+50,int(yval),ha='center',va='bottom')
```

```python
coastline = [135663, 111866, 66526, 17968, 45389]

# Set the positions and width for the bars
bar_width = 0.35
index = np.arange(len(zones))
print(index)
# Create the bar plot
plt.figure(figsize=(12, 6))
bars1 = plt.bar(index, average_depths, bar_width, label='Average Depth (m)', color='blue')
bars2 = plt.bar(index + bar_width, coastline, bar_width, label='Coastline (km)', color='red')

# Adding titles and labels
plt.title('Average Depth and Width of Ocean Zones')
plt.xlabel('Ocean Zone')
plt.ylabel('Measurements')
plt.xticks(index + bar_width / 2, zones)

# Add a legend
plt.legend()

# Add values on top of bars
for bars in [bars1, bars2]:
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, yval + 100, int(yval), ha='center', va='bottom')

# Show the plot
plt.show()
```

**Exercise 2**

```python
#exercise:  I have three fish types, with their average sizes, and average weights.
fish_types = ["Salmon", "Tuna", "Trout"]
average_sizes = [28.0, 40.0, 20.0]  # average size in inches
average_weights = [8.0, 30.0, 5.0]  # average weight in pounds
average_ages = [0.5, 1, 3]

#Make barplots in one figure showing the size, weight and age

```

```python
index=np.arange(len(fish_types))
bar_width = 0.2
bars1=plt.bar(index,average_sizes,bar_width,label='average size',color='black')
bars2=plt.bar(index+bar_width,average_weights,bar_width,label='average weights',color='blue')
bars3=plt.bar(index+bar_width*2,average_ages,bar_width,label='average ages',color='red')

plt.title('Size, weight and age of different fish')
plt.xlabel('Fish Type')
plt.ylabel('Measurements')
plt.xticks(index+bar_width,fish_types)
plt.legend()
```

**3. Scattered Plot**

**3. Scattered Plot**

```python
# Use scattered plot to look at the relationship between fish average size and average weight
plt.figure(figsize=(10,6))
plt.scatter(average_sizes,average_weights,color='r',marker='o')

#add labels and titlte
for i, fish in enumerate(fish_types):
  plt.text(average_sizes[i],average_weights[i],fish,fontsize=12)

plt.xlabel('Average Size (in inches)')
plt.ylabel('Average Weight (in pouds)')
plt.title('Scatter Plot of Fish Average Size vs Average Weight')
```

**4. Piechart**

```python
# Data: Percentage composition of different types of fish in a sample
fish_types = ['Salmon', 'Tuna', 'Trout', 'Cod', 'Herring']
percentages = [25, 20, 15, 25, 15]

colors=['red','blue','yellow','green','orange']
plt.pie(percentages, labels=fish_types,colors=colors,autopct='%1.1f%%')
plt.axis('equal') # with equal aspect ratio to make sure the pie is drawn as a circle
```

**5. Contour Plot**

```python
x, y = np.meshgrid(np.arange(7), np.arange(10))
z = np.sin(0.5 * x) * np.cos(0.52 * y)

plt.contourf(x,y,z)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
```
